# 03 - Cleaning/Wrangling/EDA

<li> Deal with NaN values </li>

```python
#check for NaN values
data.isna().values.any()
data.isna().sum()
#convert to the percentage
round(data.isna().sum()/len(data)*100,2)
```

```python
#drop the column with more than half NaN values
data.drop("vehicle_type", axis = 1, inplace = True)
```

```python
#check the columns with NaN values
nan_col = ["state", "response", "months_since_last_claim", "number_of_open_complaints", "vehicle_class", "vehicle_size"]
for i in nan_col:
    print(f"{i} - {data[i].unique()}")
```

```python
#deal with NaN values according to the data
#impute with half of the values
data.loc[data["months_since_last_claim"].isna(), "months_since_last_claim"] = data["months_since_last_claim"].quantile(0.5)
data.loc[data["number_of_open_complaints"].isna(), "number_of_open_complaints"] = data["number_of_open_complaints"].quantile(0.5)
#fill with the previous value
data = data.fillna(method = "pad")
```

# 04 - Processing Data

<li> Dealing with outliers </li>

```python
#check the distribution of numerical values and outliers
for i in data_num.columns:
    sns.distplot(data_num[i], bins=40)
    plt.show()
```

```python
#remove the outliers
#because of high correlation between monthly_premium_auto and total_claim_amount, they will not be changed
iqr = np.percentile(data_num["customer_lifetime_value"], 75)- np.percentile(data_num["customer_lifetime_value"], 25)
upper_limit = np.percentile(data_num["customer_lifetime_value"], 75) + 1.5*iqr
lower_limit = np.percentile(data_num["customer_lifetime_value"], 75) - 1.5*iqr
data_num["customer_lifetime_value"] = data_num[(data_num["customer_lifetime_value"]>lower_limit) & (data_num["customer_lifetime_value"]<upper_limit)]["customer_lifetime_value"]
```

```python
#show the result in boxplot
fig, ax = plt.subplots(4,1)

sns.boxplot(x = data_num['income'], ax = ax[0], color ='orange')
sns.boxplot(x = data_num['customer_lifetime_value'], ax = ax[1], color ='green')
sns.boxplot(x = data_num['monthly_premium_auto'], ax = ax[2], color ='red')
sns.boxplot(x = data_num['total_claim_amount'], ax = ax[3])

fig.tight_layout()
plt.show()
```

<li> Normalization </li>

```python
#check NaN values
data_num.isna().sum()
```

```python
#define a function to clean values before transformation
def transfom_clean_(x):
    if np.isfinite(x):
        return x
    else:
        return np.NAN
#apply to the features that has NaN values
data_num['customer_lifetime_value'] = list(map(transfom_clean_, data_num['customer_lifetime_value']))
#fill NaN values with the mean
data_num['customer_lifetime_value'] = data_num['customer_lifetime_value'].fillna(np.mean(data_num['customer_lifetime_value']))
                                            
```

```python
corr_matrix = data_num.corr()
sns.heatmap(round(corr_matrix, 2), annot=True)
plt.show()
```

```python
#define the target feature
y = data_num['total_claim_amount']
#drop features with min relation
data_num.drop(['months_since_last_claim', 'months_since_policy_inception',
       'number_of_open_complaints'], axis = 1, inplace = True)
```

```python
#normalize features with great values
transformer = Normalizer().fit(data_num[['income', 'customer_lifetime_value']])
X_normalized = transformer.transform(data_num[['income', 'customer_lifetime_value']])
sns.displot(X_normalized, bins = 50)
plt.show()
```

```python
#replace 0 with the mean for transformation
data_num["income"] = data_num["income"].replace(0, np.mean(data_num["income"]))
#apply logarithmic transformation to the numerical features
#drop the target value
X = data_num.drop(['total_claim_amount'], axis = 1)
log_transformer = np.log(X)
sns.displot(log_transformer, bins = 70)
plt.show()
```

<li> Encoding Categorical Data </li>

```python
data_cat.columns
```

```python
#drop irrelevant columns
to_encode = data_cat.drop(['effective_to_date','customer', 'coverage'], axis=1)
```

```python
#encode with OneHotEncoder
encoder = OneHotEncoder(drop='first').fit(to_encode)
encoded = encoder.transform(to_encode).toarray()
pd.DataFrame(encoded)
```

```python
#define the feature that will be labeled and create dataframe
label_data = pd.DataFrame(list(data_cat['coverage']), columns = ['Coverage'])
#encode with LabelEncoder
label_encoder = LabelEncoder().fit(label_data['Coverage'])
label_data['Coverage_cat'] = label_encoder.transform(label_data['Coverage'])
label_data.value_counts()
```

<li> Splitting into train set and test set </li>

```python
#check multicollinearity
corr_matrix = data_num.corr()
sns.heatmap(round(corr_matrix, 2), annot=True)
plt.show()
```

```python
#concatenate independent variables
#drop categorical column from label dataframe
label_data.drop("Coverage", axis = 1, inplace = True)
X = np.concatenate((log_transformer, encoded, label_data), axis= 1)
pd.DataFrame(X)
```

```python
#train-test split
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, random_state = 50)
```

# 05 - Modeling

```python
#apply train-test model
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```

```python
#find the optimal score for r2_score
r2 = r2_score(y_test, predictions)
r2
```

```python
#OLS method
X = sm.add_constant(X)
model = sm.OLS(y,X).fit()
model.summary()
```

# 06 - Model Validation
